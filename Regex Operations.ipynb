## 1.1 Change  the hashtags to lower case 
e.g. replace #Coronavirus with #coronavirus

Create a new column clean_text. The clean_text should have modified hashtags.

````python
# Code for 1.1

def CleanText(text):
  for t in re.findall(r'#\w+', text):
    #print(t)
    text = text.replace(t, str(t).lower())
  return text

df['clean_text'] = df['Tweet Content'].swifter.apply(CleanText)

# Let us look at the clean text column
for text in df['clean_text'].iloc[100:110]:
  print(text)
````

## 1.2 Remove RT from text in column clean_text


````python
# Code for 1.2

def RemoveRT(text):
  for t in re.findall(r'RT', text):
    #print(t)
    text = text.replace(t, '')
  return text

df['clean_text'] = df['clean_text'].swifter.apply(RemoveRT)

# Let us look at the clean text column
for text in df['clean_text'].iloc[100:110]:
  print(text)
````

## 1.3 Remove URLs and links from text in column clean_text

````python
# Code for 1.3

def RemoveURL(text):
  for t in re.findall(r'https?://[a-zA-Z.\/0-9]+', text):
    #print(t)
    text = text.replace(t, '')
  return text

df['clean_text'] = df['clean_text'].swifter.apply(RemoveURL)

# Let us look at the clean text column
for text in df['clean_text'].iloc[100:110]:
  print(text)
````

## 1.4 Removing Punctuations from text in column clean_text.
Hint: Use the following function
```python
text.translate(str.maketrans('', '', string.punctuation))
```

````python
# Code for 1.4

def RemovePunctuation(text):
  text = text.translate(str.maketrans('', '', string.punctuation))
  return text

df['clean_text'] = df['clean_text'].swifter.apply(RemovePunctuation)

# Let us look at the clean text column
for text in df['clean_text'].iloc[100:110]:
  print(text)
````

## 1.5 Extract number of Hashtags in a new column num_hashtags


````python
# Code for 1.5

def CountHashTags(text):
  return len(re.findall(r'#\w+', text))

df['num_hashtags'] = df['Tweet Content'].swifter.apply(CountHashTags)

# Let us look at the num_hashtags column
df['num_hashtags'].iloc[100:110]
````

## 1.6 Extract number of user mentions in a new column num_mentions

````python
# Code for 1.6

def CountMentions(text):
  return len(re.findall(r'@\w+', text))

df['num_mentions'] = df['Tweet Content'].swifter.apply(CountMentions)

# Let us look at the nums_mentions column
for text in df['Tweet Content'].iloc[100:110]:
  print(text)
  
df['num_mentions'].iloc[100:110]
````

## 1.7 Count number of mentions and hashtags

````python
# Code for 1.7

df['total_mentions_tags'] = df['num_mentions'] + df['num_hashtags']

df['total_mentions_tags'].iloc[100:110]
````

## Task 2.2 Extract image names from files names

The part after `images` in file names represent the image names. For example  for the file name `oxford-iiit-pet/images/newfoundland_61.jpg`, image name is `newfoundland_61.jpg`

Extract only image names from filenames and store them in a separate column.



````python
# Code here for 2.2

def GetImageName(text):
  #text = text[text.find('images/') + 7:]
  text = re.sub(r'.+/images/', '', text)
  return text

df['image_names'] = df['filename'].swifter.apply(GetImageName)

df['image_names'].head()
````

## Task 2.3  Extract label names from image names
Each image correspond to image of a unique breed of a dog or a cat. The ML task would be to predict the breed of cat/dog given the image file. To do this task we first need to get the outcome (label or dependent varible) from the file names. In this dataset the part before the _number.jpg represents the breed name. For example for image name `german_shorthaired_151.jpg` the breed name is erman_shorthaired. Keep in mind that same breed names are single words and some are multiple words separated by underscores.

Extract the label names as new column in teh dataframe.

````python
def extract_label(x):
  result = re.match(r'^(\w+)_\d+\.jpg', x)
  if result is not None:
    label = result.group(1)
  else:
    label = 'Error' 
  return label
````

````python
# code here for 2.3

df['label'] = df['image_names'].swifter.apply(extract_label)
````

## Task 2.4 Check whether breed name is a Cat or Dog
If the first alphabet  of the breed name is a capital letter then it is a cat else it is a dog. For example `Siamese` is a Cat whereas `chihuahua` is a dog.

Add a new column `Cat' which takes the value of $1$ if breed corresponds t0 a cat and $0$ if it corresponds to a dog.

````python
def IsCat(ipText):
  return 1 if ipText[0].isupper() else 0
````

````python
# code here
df['Cat'] = df['image_names'].swifter.apply(IsCat) 
````
